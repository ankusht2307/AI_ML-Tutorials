# DATA CLEANING AND PREPARATIONS

```python
import pandas as pd

# Sample DataFrame with missing values and duplicates
data = {'A': [1, 1, None, 4, 5],
        'B': [10, 10, 30, 40, 50],
        'C': [1, 1, 3, 2, 1]}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
# Handling missing values

# Drop rows with missing values
df_cleaned = df.dropna()
print("\nDataFrame after handling missing values (dropping):")
print(df_cleaned)

# Fill missing values with mean
df_imputed = df.fillna(df.mean())
print("\nDataFrame after handling missing values (imputation):")
print(df_imputed)
```

```python
# Removing duplicates
df_no_duplicates = df_imputed.drop_duplicates()
print("\nDataFrame after removing duplicates:")
print(df_no_duplicates)
```

```python
# Data normalization
# Min-Max normalization
df['B_normalized'] = (df['B'] - df['B'].min()) / (df['B'].max() - df['B'].min())
print("\nDataFrame after data transformation and normalization:")
print(df)

# 30-10 =20 / 50-10 = 40 
```

Is there a correlation between sales and profit?

What is the average discount by region?

```python
# Question 1: Total sales amount for each product category
total_sales_by_category = df.groupby('Product Category')['Sales'].sum()
print("\nTotal sales amount for each product category:")
print(total_sales_by_category)

# Question 2: Correlation between sales and profit
sales_profit_corr = df['Sales'].corr(df['Profit'])
print("\nCorrelation between sales and profit:", sales_profit_corr)

# Question 3: Average discount by region
average_discount_by_region = df.groupby('Region')['Discount'].mean()
print("\nAverage discount by region:")
print(average_discount_by_region)
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 30, 35] 

# Sort categories and values in descending order
sorted_data = sorted(zip(values, categories), reverse=True)
sorted_values, sorted_categories = zip(*sorted_data)

# Create bar plot
plt.figure(figsize=(8, 6))
plt.bar(sorted_categories, sorted_values, color='skyblue')

# Best Practices
plt.title('Bar Graph of Categories (Descending Order)')
plt.xlabel('Categories')
plt.ylabel('Values')

# Labeling
for i in range(len(sorted_categories)):
    plt.text(i, sorted_values[i] + 0.5, str(sorted_values[i]), ha='center', va='top')

# Adjusting Spacing
plt.tight_layout()

# Show plot
plt.show()
```

# FEATURE SELECTION 

```python
# Import necessary libraries
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target

# Initialize Random Forest Classifier
rf = RandomForestClassifier()

# Fit model
rf.fit(X, y)

# Get feature importances
feature_importances = pd.DataFrame(rf.feature_importances_,
                                   index = iris.feature_names,
                                   columns=['importance']).sort_values('importance', ascending=False)

# Print feature importances
print(feature_importances)
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import LabelEncoder

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target

# Convert target to categorical for IV calculation
label_encoder = LabelEncoder()
y_cat = label_encoder.fit_transform(y)

# Create a DataFrame to hold the dataset
df = pd.DataFrame(X, columns=iris.feature_names)
df['target'] = y_cat

# Function to calculate IV for a feature
def calculate_iv(df, feature, target):
    iv = 0
    for category in df[feature].unique():
        good = df[(df[feature] == category) & (df[target] == 1)].shape[0]
        bad = df[(df[feature] == category) & (df[target] == 0)].shape[0]
        total_good = df[df[target] == 1].shape[0]
        total_bad = df[df[target] == 0].shape[0]
        if good != 0 and bad != 0:
            iv += (good / total_good - bad / total_bad) * (pd.np.log(good / total_good) - pd.np.log(bad / total_bad))
    return iv

# Calculate IV for each feature
iv_values = {}
for col in df.columns[:-1]:  # Exclude the target column
    iv_values[col] = calculate_iv(df, col, 'target')

# Print IV values
print("Information Value (IV) for each feature:")
for feature, iv in iv_values.items():
    print(f"{feature}: {iv:.4f}")
```
